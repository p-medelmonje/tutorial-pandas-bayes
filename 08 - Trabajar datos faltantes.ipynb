### Eliminar filas con NaN

```python
""" .dropna() elimina filas o columnas que contengan datos faltantes. El argumento axis se usa para especificar
el eje, siendo 0 para filas y 1 para columnas. how = "all" elimina todas las filas que contenga todos sus
datos faltantes"""

titanic1 = titanic.dropna(axis = 0, how = "all")
titanic1.shape

# ninguna fila tiene todos sus datos faltantes. Ninguna se eliminó
```

```python
# El argumento how = "any" elimina todas las filas que tengan al menos un valor faltante

titanic2 = titanic.dropna(axis = 0, how = "any")
titanic2.shape
```

### Rellenar NaN

```python
# Contar los registros faltantes dentro de la columna "age"

pd.isnull(titanic["age"]).values.ravel().sum()
```

```python
# Para rellenar los valores faltantes con el promedio de los valores que se tienen:

col = titanic["age"].fillna((titanic["age"]).mean())
```

```python
col.head(50)
```

```python
# Crear una nueva columna en el dataset original

titanic1["new_age"] = col
titanic1.head(17)
```

```python
# Rehacer dataset reemplazando la columna "age" por la columna "new_age"

titanic1 = titanic1[["pclass", "survived", "name", "sex", "new_age", "sibsp", "parch", "ticket",
                     "fare", "cabin", "embarked", "boat", "body", "home.dest"]]

titanic1.head(30)
```
